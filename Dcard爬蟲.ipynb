```python
import pandas as pd
import requests
import re
from fake_useragent import UserAgent
import random
import time
import cloudscraper
ua = UserAgent()
```

```python
scraper = cloudscraper.create_scraper()
```

```python
def Crawl(ID):
    link = 'https://www.dcard.tw/service/api/v2/posts/' + str(ID)
    scrp = scraper.get(link, headers = {
    "user-agent": ua.random
  })
    scrpjs = scrp.json()
    return(pd.DataFrame(data=
        [{'id':scrpjs['id'],
          '標題':scrpjs['title'],
          '摘要':scrpjs['excerpt'],
          '校名匿名':scrpjs['anonymousSchool'],
          '匿名':scrpjs['anonymousDepartment'],
          '發文時間':scrpjs['createdAt'],
          '更新時間':scrpjs['updatedAt'],
          '回應數量':scrpjs['commentCount'],
          '愛心數量':scrpjs['likeCount'],
          '標註':scrpjs['topics'],
          '分類':scrpjs['forumName'],
          '性別':scrpjs['gender'],
          '心情細項':scrpjs['reactions'],
          '留言數量':scrpjs['commentCount'],
          '貼文回應':scrpjs['reactions']}],
        columns=['id','標題','摘要','校名匿名','匿名','發文時間','更新時間','回應數量','愛心數量','標註','分類','性別','心情細項','留言數量','貼文回應']))
```

# 爬取文章

```python
# 一次讀取100篇最熱門的文章
## 跑一次就好 ##
url = 'https://www.dcard.tw/service/api/v2/forums/relationship/posts?popular=false&limit=100'
resq = scraper.get(url)
rejs = resq.json()
df = pd.DataFrame()
for i in range(len(rejs)):
  df = df.append(rejs[i],ignore_index=True)
print(df.shape)
```

```python
# 透過迴圈讀取10*100篇文章，若需讀取更多資料，可以將range(10)中的數值提升
k=0
for j in range(100):
  k+=1
  time.sleep(15)
  last = str(int(df.tail(1).id)) # 找出爬出資料的最後一筆ID
  url = 'https://www.dcard.tw/service/api/v2/forums/relationship/posts?popular=false&limit=100&before=' + last
  resq = scraper.get(url)
  rejs = resq.json()
  print(k)
  for i in range(len(rejs)):
    df = df.append(rejs[i], ignore_index=True)
    
print(df.shape)

df.id = df.id.astype(int)
```
